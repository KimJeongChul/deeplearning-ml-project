```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x, y = np.meshgrid(range(10), range(10))
z_1 = (-1.2 * x - 0.7 * y + 1.0)
z_2 = (-1.2 * x - 0.7 * y + 2.0)
```

```python
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(x, y, z_1)
plt3d.plot_surface(x, y, z_2)
plt.show()
```
